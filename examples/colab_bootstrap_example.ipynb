# mktools bootstrap in Colab

This notebook shows the recommended pattern:

1. Install `mktools`
2. Import `mktools.common.bootstrap`
3. Run one bootstrap call
4. Use `mktools.kio`, `mktools.kfile`, and `mktools.kstat`

The public `mktools` repo should hold the generic bootstrap logic.
Private path conventions and secrets should live in a private overlay JSON file or in environment variables.

```python
from mktools.common.bootstrap import (
    BootstrapSettings,
    FrameworkMode,
    SecretPolicy,
    bootstrap_environment,
    print_bootstrap_summary,
)
```

```python
# Optional private overlay path.
# Keep the file private or gitignored if it contains personal folder structure.
OVERLAY_JSON = "/content/drive/MyDrive/private/bootstrap_overlay.json"

settings = BootstrapSettings(
    project_name="Capstone",
    runlevel="development",
    analysis_depth="deep",
    framework_mode=FrameworkMode.TENSORFLOW,
    random_seed=42,
    mount_drive=True,
    configure_inline_matplotlib=True,
    configure_seaborn_theme=True,
    configure_kaggle=False,
    kaggle_policy=SecretPolicy.OPTIONAL,
    overlay_json_path=OVERLAY_JSON,   # set to None if not using an overlay
    install_requirements=[],
)

boot = bootstrap_environment(settings)
print_bootstrap_summary(boot)
```

```python
# Typical mktools workflow after bootstrap
from mktools.kio import DataImporter
from mktools.kstat import KStatProfiler
from mktools.kfile import DirectoryScanner

print("Raw data path :", boot.paths.data_raw)
print("Reports path  :", boot.paths.reports_dir)
print("Workspace path:", boot.paths.project_workspace)
```
